## 데이터 Augmentation 정답데이터 만드는 법

### 1. 원본데이터 로드

```python
import pandas as pd


import os

# D drive
data_path = 'D:\dataset\Dacon\dirty_mnist/'

data = pd.read_csv(data_path + 'dirty_mnist_2nd_answer.csv')
print(data.shape)
data.head()
```

### 2. 데이터 이어붙이기
- Pandas dataframe append()

```python
# 원본 + 90도 rotation (== 270도 rotation)
data2 = data.append(data)
print(data2.shape)

data3 = data.append(data).append(data)
print(data3.shape)
```

## 3. index 기준 정렬 및 to_csv

```python
data2 = data2.sort_values(by='index')
data2.to_csv('dirty_mnist_2nd_answer_90_or_270.csv', index=False)

data3 = data3.sort_values(by='index')
data3.to_csv('dirty_mnist_2nd_answer_90_and_270.csv', index=False)

print('to_csv완료')
```
